```python
change_col=["Workclass","Occupation","Native-country"]

for col in change_col:
    data_copy[col]=data_copy[col].str.replace("?","Unknown")
    print(data_copy[col].value_counts(),"\n",data_copy[col].unique())
    print("_______________________")
```

```python
unique_educ=["Education","Education_num"]

for edu in unique_educ:
    print(data_copy[edu].value_counts(),"\n",data_copy[edu].unique())
    print("----------------")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/veena9026/stats-assignment/main/adult_clean_data.csv")
```

```python
df.head(6)
```

```python
df["Education_num"].value_counts().plot.pie(y=df["Education"],figsize=(20,20),autopct='%1.1f%%')
```

```python
plt.figure(figsize=(25,10))
sns.scatterplot(y="Age",x="Occupation",data=df)
```

```python
plt.figure(figsize=(30,10))
sns.histplot(x="Occupation",data=df)
```

```python
plt.figure(figsize=(25,10))
sns.boxplot(x="Age",y="Sex",data=df)
plt.xticks(size=15,rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.barplot(y="Age",x="Occupation",data=df)
plt.xticks(size=15,rotation=90)
plt.show()
```

```python
sns.barplot(x="Education_num",y="Workclass",data=df)
```

```python
# Creating a donut chart for 'Age'
relation = df['Relationship'].value_counts()

plt.style.use('bmh')
plt.figure(figsize=(20, 10))
plt.pie(relation.values, labels=relation.index,
        startangle=50, autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Relationship distribution', fontdict={
          'fontname': 'Monospace', 'fontsize': 30, 'fontweight': 'bold'})
plt.axis('equal')
plt.legend(prop={'size': 15})
plt.show()
```

```python
numerical_feature=[feature for feature in df.columns if df[feature].dtypes !="O"]
numerical_feature
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numerical_feature)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numerical_feature[i]],shade=True, color='r')
    plt.xlabel(numerical_feature[i])
    plt.tight_layout()
```

```python
categorical_feature=[feature for feature in df.columns if df[feature].dtypes =="O"]
categorical_feature
```

```python
# categorical columns
plt.figure(figsize=(60, 60))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=50, fontweight='bold', alpha=0.8, y=1.)
category =  categorical_feature
for i in range(0, len(category)):
    plt.subplot(2, 5, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=90)
    plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,5))
sns.pairplot(df, hue='Annual_income', height=2)
plt.show()
```

```python
# Clustered Bar Chart 
plt.figure(figsize=[8,6])
ax = sns.barplot(data = df, x = 'Annual_income', y = 'Age', hue = 'Sex')
ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'Sex')
plt.title('Average of Age for Sex relative to Income')
plt.xlabel('Income ($)')
plt.ylabel('Average of Age');
```

```python
# Bar Chart 
plt.figure(figsize=[8,6])
sns.barplot(data=df, x='Annual_income', y='Hours-per-week', palette='YlGnBu')
plt.title('Average of Hours per Week relative to Income')
plt.xlabel('Income ($)')
plt.ylabel('Average of Hours per Week');
```

```python
# Creating a barplot for 'Hours per week'
hours = df['Hours-per-week'].value_counts().head(10)

plt.style.use('bmh')
plt.figure(figsize=(15, 7))
sns.barplot(hours.index, hours.values, palette='colorblind')
plt.title('Distribution of Hours of work per week', fontdict={
          'fontname': 'Monospace', 'fontsize': 20, 'fontweight': 'bold'})
plt.xlabel('Hours of work', fontdict={'fontname': 'Monospace', 'fontsize': 15})
plt.ylabel('Number of people', fontdict={
           'fontname': 'Monospace', 'fontsize': 15})
plt.tick_params(labelsize=12)
plt.show()
```
