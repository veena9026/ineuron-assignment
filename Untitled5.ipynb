```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab


def plot_data (Sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(Sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(Sample,dist="norm",plot=pylab)
    plt.show
```

```python
plot_data(np.log(p))
```
