## EDA


```python
df=data.copy()
```

```python
df.head()
```

```python
df.groupby(["Page_Rank"])["Google_Index"].value_counts().plot.pie(figsize=(8,8),autopct='%1.1f%%')
```

```python
sns.barplot(x="popUpWidnow",y="Google_Index",data=df,hue="Result")

```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="having_IP_Address",y="URL_Length",hue="Result",data=df)
plt.xticks(size=15,rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x="URL_Length",hue="Result",data=df)
```

```python
fig,axs=plt.subplots(1,4,figsize=(25,25))
plt.subplot(411)
sns.countplot(data=df,x="Shortining_Service",hue="Result",palette="pastel")
plt.subplot(412)
sns.countplot(data=df,x="having_At_Symbol",hue="Result",palette="ch:.25")
plt.subplot(413)
sns.countplot(data=df,x="double_slash_redirecting",hue="Result",palette="flare")
plt.subplot(414)
sns.countplot(data=df,x="Domain_registeration_length",hue="Result",palette="Set2")
```

```python
for i in df.columns:
    label_size=dict(df[i].value_counts())
    label=list(label_size.keys())
    size=list(label_size.values())
    plt.pie(size,labels=label,autopct="%1.1f%%")
    plt.title(f"Result vs {i}")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
